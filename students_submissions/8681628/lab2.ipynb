# 1. Line plot  using sample data

```python
import seaborn as sns
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 18, 22]

sns.lineplot(x=x, y=y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot with Seaborn')
plt.show()
```

# 2. Bar plot  using sample data

```python
import seaborn as sns
import matplotlib.pyplot as plt

categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [30, 45, 22, 60]

sns.barplot(x=categories, y=values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot with Seaborn')
plt.show()
```

# 3. Scatter plot  using sample data

```python
import seaborn as sns
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 18, 22]

sns.scatterplot(x=x, y=y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Seaborn')
plt.show()
```

# 4. Histogram plot  using sample data

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)

sns.histplot(data, bins=30, color='blue', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram with Seaborn')
plt.show()
```

# 5. Box plot  using iris dataset with sepal length and species on y and x axes 

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset('iris')

sns.boxplot(x='species', y='sepal_length', data=data)
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.title('Box Plot with Seaborn')
plt.show()
```

# 6. Pair plot  using iris dataset showing sepal and petal's length and width

```python
import seaborn as sns

data = sns.load_dataset('iris')

sns.pairplot(data, hue='species')
plt.title('Pair Plot with Seaborn')
plt.show()
```

# 7. Violin plot  using  tips dataset showing total bill value on each day of the week.

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset('tips')

sns.violinplot(x='day', y='total_bill', data=data, inner='stick')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.title('Violin Plot with Seaborn')
plt.show()
```
